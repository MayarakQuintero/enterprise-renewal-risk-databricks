```sql
-- If we intervene on the top 10%, 20%, 30% accounts, how many churners do we catch
SELECT
  CAST(top_percent * 100 AS INT) AS top_percent,
  k_accounts,
  true_churners_captured,
  ROUND(precision_at_k, 3) AS precision_at_k,
  ROUND(recall_at_k, 3) AS recall_at_k
FROM renewal_model_topk_eval
ORDER BY top_percent
```
