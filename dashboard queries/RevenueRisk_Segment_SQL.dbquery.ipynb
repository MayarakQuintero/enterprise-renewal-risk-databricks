```sql
SELECT
  industry,
  COUNT(*) AS accounts,
  ROUND(SUM(revenue_at_risk), 0) AS revenue_at_risk,
  ROUND(AVG(non_renewal_probability), 3) AS avg_risk_score
FROM renewal_dashboard_base
GROUP BY industry
ORDER BY revenue_at_risk DESC;

SELECT
  region,
  COUNT(*) AS accounts,
  ROUND(SUM(revenue_at_risk), 0) AS revenue_at_risk,
  ROUND(AVG(non_renewal_probability), 3) AS avg_risk_score
FROM renewal_dashboard_base
GROUP BY region
ORDER BY revenue_at_risk DESC;

SELECT
  company_size,
  COUNT(*) AS accounts,
  ROUND(SUM(revenue_at_risk), 0) AS revenue_at_risk,
  ROUND(AVG(non_renewal_probability), 3) AS avg_risk_score
FROM renewal_dashboard_base
GROUP BY company_size
ORDER BY revenue_at_risk DESC;
```
