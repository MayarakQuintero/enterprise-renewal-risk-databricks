```sql
SELECT
  risk_rank,
  client_id,
  industry,
  region,
  company_size,
  contract_end_date,
  ROUND(non_renewal_probability, 4) AS non_renewal_probability,
  ROUND(annual_contract_value, 2) AS annual_contract_value,
  ROUND(revenue_at_risk, 2) AS revenue_at_risk,
  top_driver_summary,
  recommended_action
FROM renewal_dashboard_base
ORDER BY risk_rank
LIMIT 5;
```
